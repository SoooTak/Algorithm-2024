#### 1) 문제정의
이진트리가 주어졌을 때 트리의 높이를 구하는 방법을 분할 정복 전략을 통해 해결해보자

#### 2)알고리즘 설명
알고리즘의 입력은 노드 root로 어떤 르티의 루트를 나타낸다.
만약 루트가 None이면 공백 트리를 의미하므로 높이 0을 바로 반환한다. 
공백 트리가 아니면 두 서브 트리의 높이를 계산하며, 순환 호출을 사용한다.
마지막으로 결과를 비교해 더 큰 값에 1을 더한 값을 반환한다.

#### 4) 알고리즘 코드 개요
함수명 : calc_height(root):

매개 변수 : hLeft, hRight(트리의 왼쪽, 오른쪽 높이)

출력 : max(hLeft,hRight)+1

수행순서:

    1. 공백 트리이면 0을 반환

    2. 높이가 0 -> 0을 반환

    3. 왼쪽 트리의 높이 및 오른쪽 트리의 높이 반환

    4. 더 높은 높이에 1을 더해 반환

#### 5) 알고리즘 코드


```python
def calc_height(root):
    if root is None:
        return 0
    hLeft = calc_height(root.left)
    hRight = calc_height(root.right)

    return max(hLeft,hRight)+1
```
