### 2) 알고리즘 설명  (MD)

거리 변수에 임의의 높은 값을 저장해둔 뒤, 이중 반복문을 사용하여 점 사이의 거리를 구해 최저 거리를 갱신해서 최종적으로 가장 작은 거리를 반환한다.

### 4) 알고리즘 코드 개요(입력 변수, 출력, 알고리즘 개요) (MD)

함수 명 : closest_pair

매개변수 : 좌표(x, y) 리스트 p

(예시) p = [(1,1), (2,2), (5,5)]

출력 : 거리 변수 mindlist (최근접 거리)

수행 순서 :

1. 점의 개수를 변수 n에 저장한다. n = len(p)
2. 거리 변수 mindlist에 임의의 큰 값(무한대로 지정)을 저장한다. mindlist = float("inf")
3. 중복 거리 계산을 방지하기 위해 이중 for문을 통해 i < j를 만족하는 동안 모든 쌍에 대한 거리를 계산해 mindlist와 비교해서 최근접 거리를 갱신한다.
4. 최종적으로 mindlist에 최근접 거리가 저장되어 반환된다.

### 5) 알고리즘 코드 (PY,C)

```python
import math

def distance(a, b) :
    return math.sqrt((a[0] - b[0]) ** 2 + (a[1] - b[1]) ** 2)
    
def closest_pair(p) :
    n = len(p)
    mindlist = float("inf")
    for i in range(n-1) :
        for j in range(i+1, n) :
            dist = distance(p[i], p[j])
            if dist < mindlist :
                mindlist = dist
    return mindlist
```

###   6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
import math
    
list = [(1,1), (2,2), (5,5), (5,7), (7,7)]
print(closest_pair(list))
```
