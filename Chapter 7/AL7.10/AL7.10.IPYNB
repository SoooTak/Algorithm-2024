#### 5. 코드(PY,C)

```python
import copy
def shortest_path_floyd(vertex,W):
    vsize = len(vertex)
    D=copy.deepcopy(W)

for k in range(vsize):
    for i in range(vsize):
        for j in range(vsize):
            if(D[i][k]+D[k][j]<D[i][j]):
                D[i][j]=D[i][k]+D[k][j]
    printD(D)
```

Floyd의 알고리즘은 3중 루프로 이루어져 있고, 각 루프의 최대 n번 반복한다. 따라서 시간 복잡도 O(n^3)이다.
