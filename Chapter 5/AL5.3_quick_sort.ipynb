### 2) 알고리즘 설명
리스트에서 피벗을 선택하고 이를 기준으로 리스트를 2개의 부분 리스트로 나누며, 각 부분 리스트에서 다시 같은 분할 과정을 적용한다. 이러한 분할과정은 부분 리스트를 더 이상이 나눌 수 없을 때 까지 반복된다. 부분 리스트의 크기가 1면 분할이 종료

### 4) 알고리즘 코드 개요
함수명 : quick_sort(A,lefrt,right):

매개변수: 입력리스트 data

출력 :  정렬된 data

수행순서:
    1. 정렬할 항목이 2개 이상인 경우에 대해서만 처리, 1일 경우 이미 정렬
    2. 분할 알고리즘을 이용해 리스트를 좌우로 분할
    3. 왼쪽 부분 리스트와 오른쪽 부분 리스트를 각각 정렬한다.


### 5) 알고리즘 코드


```python
def quick_sort(A,left,right):
    if left < right:
        mid = partition(A,left,right)
        quick_sort(A,left,mid-1)
        quick_sort(A,mid+1,right)
```

### 6) 테스트 코드


```python
def partition(A, left, right):
    low = left +1
    high = right
    pivot = A[left]
    while low <= high:
        while low <= right and A[low] <= pivot: low += 1
        while high >= left+1 and A[high] > pivot: high -= 1
        if low < high:
            A[low], A[high] = A[high], A[low]
    A[left], A[high] = A[high], A[left] # 이 부분을 while 루프 밖으로 옮깁니다.
    return high

def quick_sort(A,left,right):
    if left < right:
        mid = partition(A,left,right)
        quick_sort(A,left,mid-1)
        quick_sort(A,mid+1,right)
    
data = [5,3,8,4,9,1,6,2,7]
sorted = [0]*len(data)
print("Oringinal : ",data)
quick_sort(data,0,len(data)-1)
print("QuickSort :",data)
```
