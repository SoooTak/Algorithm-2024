### 4) 알고리즘 코드 개요(입력 변수, 출력, 알고리즘 개요)
함수명 : merge_sort(A, left, mid, right):
매개 변수 : 임시리스트 k, 왼쪽 리스트  인덱스 i, 오른쪽 리스트 인덱스 j
출력 : 정렬된 리스트 A
수행순서 :
    1. 정렬할 항목이 2개 이상이 경우에 대해서만 분할-정복을 진행
    2. 리스트를 균등하게 불할할 인덱스를 계산한다.
    3. 왼쪽 부분리스트와 오른쪽 부분 리스트를 각각 독집적으로 병할 정렬
    4. 정렬된 두 리스트를 병합.
    

### 5) 알고리즘 코드 (PY,C)


```python
def merge_sort(A,left,right):
    if left<right :
        mid = (left+right) //2
        merge_sort(A,left,mid)
        merge_sort(A,mid+1,right)
        merge(A,left,mid,right)
```

### 6) 테스트 코드(PY,C)

```python
def merge_sort(A,left,right):
    if left<right :
        mid = (left+right) //2
        merge_sort(A,left,mid)
        merge_sort(A,mid+1,right)
        merge(A,left,mid,right)

def merge(A,left,mid,right):
    k = left
    i = left
    j = mid +1
    while i <= mid and j <=right :
        if A[i] <= A[j]:
            sorted[k]=A[i]
            i,k=i+1, k+1
        else:
            sorted[k]=A[j]
            j,k=j+1,k+1

    if i>mid :
        sorted[k:k+right-j+1]=A[j:right+1]
    else :
        sorted[k:k+mid-i+i]=A[i:mid+1]

    A[left:right]= sorted[left:right]
    
data = [5,3,8,4,9,1,6,2,7]
sorted = [0]*len(data)
print("Oringinal : ",data)
merge_sort(data,0,len(data)-1)
print("Mergesort :",data)
```

### 8) 복잡도 분석
병합 정렬은 순환 호출 구조로 되어 있다. 따라서 레코드의 개수 n이 2의 거듭제곱이라고 가정하고 순환 호출의 깊이가 얼마나 되는지를 분석해 보자. n=2^3인 경우에는 부분배열의 크기가 2^3 -> 2^2 -> 2^1 -> 2^0  순으로 줄어들어 순환 호출의 길이가 3임을 알 수 있다. 따라서 n=2^k라고 하면 부분배열의 크기는 2^k -> 2^k-1 ...2^0이 되어 순환 호출의 깊이는 k 이때 k = log2n
