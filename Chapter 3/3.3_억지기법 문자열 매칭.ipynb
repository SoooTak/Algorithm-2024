### 1)문제 정의
문제 3.3 문자열 매칭 문제
길이가 n인 입력 문자열 T와 길이가 m인 패턴 문자열 P가 있다. T에서 가장 먼저 나타나는 P의 위치를 찾아라. 패턴이 없으면 -1을 반환하라 

### 2) 알고리즘 설명(MD)
텍스트 문자와 패턴문자를 입력 받은 후 텍스트 문자 첫번째 값과 패턴 문자의 값을 비교한 후 패턴 첫문자와 불일치시 패턴 전체를 다음 텍스트 문자 위치로 이동 후 다시 텍스트 문자와 패턴 문자를 비교하여 매칭 성공시 다음 패턴을 비교하여 패턴이 일치하면 위치를 반환한다.

### 3) 손으로 푼 예제 (MD, 이미지 삽입)


### 4) 알고리즘 코드 개요(입력 변수, 출력, 알고리즘 개요) (MD)

함수 명 : string_matching(T, P):

매개변수 : 입력 문자열 T, 패턴 문자열 P 문자열 T,P


출력 : 문자 인덱스 위치 i  또는 return -1(매칭 실패) 

수행 순서 :

1. 문자열을 T에 입력, 패턴 문자열을 P에 입력한다.
2. 문자열 T와 문자열 P의 문자열 길이를 구한다.(반복횟수를 구하기)
3. 문자열 T 처음 인덱스에 패턴 문자열을 모두 비교한다
4. 모든 문자가 일치하면, 현재 위치 반환 일치하지 않으면 매칭 오류를 출력

### 5) 알고리즘 코드 (PY,C)

```python
def string_matching(T,P):           #T는 입력, P는 패턴
    n = len(T)                      #n: 입력 문자열의 길이
    m = len(P)                      #m: 패턴 문자열의 길이
    for i in range(n-m+1):          #i : 0,1,... n-m
        j=0                         
        while j<m and P[j]==T[i+j]: #패턴 문자열을 모두 비교
            j=j+1                   
        if j == m:                  #모든 문자가 일치하면, 매칭 성공
         return i                   #현재 위치 반환
    return -1                       #문자열 매칭 실패
```

###   6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
def string_matching(T, P):
    n = len(T)
    m = len(P)
    for i in range(n - m + 1):
        j = 0
        while j < m and P[j] == T[i + j]:
            j += 1
        if j == m:
            return i
    return -1

# 입력 문자열과 패턴 문자열을 받습니다.
T = input("입력 문자열: ")
P = input("패턴 문자열: ")

# 문자열 매칭 함수를 호출하여 결과를 출력합니다.
index_site = string_matching(T, P)
if index_site != -1:
    print("패턴이 문자열 내에서 위치한 인덱스:", index_site)
else:
    print("패턴을 찾을 수 없다.")
```
