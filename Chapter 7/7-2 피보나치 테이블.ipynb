문제 예시 : n = 8일 때 피보나치수열의 8번째 항 21이 반환되어야 함.

### 2) 알고리즘 설명  (MD)

n+1개의 테이블을 만들어 각 테이블을 순차적으로 채워나가면서 부분문제들을 해결하는 과정을 통해 n번째 자리를 계산함.

### 3) 손으로 푼 예제 (MD, 이미지 삽입)


### 4) 알고리즘 코드 개요(입력 변수, 출력, 알고리즘 개요) (MD)

함수 이름 : fib_dp_tab(n)

매개변수 : 수열의 n번째 자리(인덱스)

출력 : 피보나치수열의 n번째 자리(값)

수행 순서 : 
1. n+1개의 테이블(f)을 만든다.
2. f[0] = 0, f[1] = 1로 초기 설정한 상태에서 반복문을 통해 f[2], f[3], f[4], ... f[n]까지 순차적으로 테이블을 채워나간다.
3. 최종적으로 f[n]을 반환한다.

### 5) 알고리즘 코드 (PY,C)

```python
#202311388 김민상
def fib_dp_tab(n) :
    f = [None] * (n+1)
    f[0] = 0
    f[1] = 1
    for i in range(2, n+1) :
        f[i] = f[i-1] + f[i-2]
    return f[n]
```

###   6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
n = int(input("피보나치수의 n번째 자리 : "))
print(fib_dp_tab(n))
```
