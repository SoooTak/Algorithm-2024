문제 예시 : listA = [8, 1, 7, 3, 9, 2, 4, 5] 가 주어지면 1, 2, 3, 4, 5, 6, 7, 8 리스트가 반환되어야 함.

### 2) 알고리즘 설명  (MD)

리스트를 항목의 개수가 하나가 될 때까지 반으로 나눠가고, 정복되면 병합을 계속하여 정렬된 전체 리스트를 완성한다.

### 4) 알고리즘 코드 개요(입력 변수, 출력, 알고리즘 개요) (MD)

함수 이름 : merge_sort(A, left, right)

매개변수 : 리스트 A, 정렬 범위 left, right

(예시) : merge_sort(listA = [8, 1, 7, 3, 9, 2, 4, 5], left, right)

출력 : 정렬된 리스트 A

수행 순서 : 
1. ((분할된)리스트의 요소가 2개 이상인 동안 : mid를 구해 범위를 둘로 나눠 리스트 분할)
2. 모두 분할(정복)되면 병합하는 과정을 거쳐 하나의 정렬된 리스트를 반환함.

### 5) 알고리즘 코드 (PY,C)

```python
def merge_sort(A, left, right) :
    if left < right :
        mid = (left + right) // 2
        merge_sort(A, left, mid)
        merge_sort(A, mid + 1, right)
        merge(A, left, mid, right)

def merge(A, left, mid, right) :
    k = left
    i = left
    j = mid + 1
    while i <= mid and j <= right :
        if A[i] <= A[j] :
            sorted[k] = A[i]
            i, k = i + 1, k + 1
        else :
            sorted[k] = A[j]
            j, k = j+1, k+1
    
    if i > mid :
        sorted[k : k+right-j+1] = A[j : right + 1]
    else :
        sorted[k : k+mid - i+1] = A[i : mid + 1]
    
    A[left : right + 1] = sorted[left : right+1]
```

###   6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
data = [5, 3, 8, 4, 9, 1, 6, 2, 7]
listA = [8, 1, 7, 3, 9, 2, 4, 5]
sorted = [0] * len(data)
print("Original : ", data)
merge_sort(data, 0, len(data) - 1)
print("MergeSort : ", data)
print()
print("Original : ", listA)
merge_sort(listA, 0, len(listA) - 1)
print("MergeSort : ", listA)
```
