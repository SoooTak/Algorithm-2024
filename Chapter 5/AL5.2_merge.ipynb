### 2) 알고리즘 설명
두 부부 리스트의 항목 A[i]와 A[j]를 비교하여 더 작은 요소를 먼저 새로운 리스트 sorted[k]로 복사한다.
이 과정은 리스트의 하나의 모든 항목이 복사될 때까지 반복된다. 마지막으로 다른 리스트의 남은 항목들을 모두 c로 복사하면 병합이 종료된다.

### 4) 알고리즘 코드 개요(입력 변수, 출력, 알고리즘 개요)

함수명 : merge(A, left, mid, right):

매개 변수 : 임시리스트 k, 왼쪽 리스트  인덱스 i, 오른쪽 리스트 인덱스 j

출력 : 정렬된 리스트 A

수행순서 :

    1. 각각 정렬된 왼쪽 부분 리스트 A[left..mid]와 오른쪽 부분 리스트 A[mid+1..right]를 하나의 리스트 A[left..fight]

    2. 인덱스 초기화 및 한 쪽의 부분 리스트가 처리될 때까지 반복
    
    3. 한쪽의 부분 리스트에 남아 있는 모든 레코드를 임시 리스트 sorted로 일괄 복사한다.

    4. 병합이 끝나면 병합 결과를 저장한 임시 리스트 sorted를 반드시 원래의 리스트 A에 복사한다. 

### 5) 알고리즘 코드 (PY,C)


```python
def merge(A,left,mid,right):
    k = left
    i = left
    j = mid +1
    while i <= mid and j <=right :
        if A[i] <= A[j]:
            sorted[k]=A[i]
            i,k=i+1, k+1
        else:
            sorted[k]=A[j]
            j,k=j+1,k+1

    if i>mid :
        sorted[k:k+right-j+1]=A[j:right+1]
    else :
        sorted[k:k+mid-i+i]=A[i:mid+1]

    A[left:right]= sorted[left:right]
    
```

### 6) 테스트 코드(PY,C)

### 8) 복잡도 분석
병합 정렬은 순환 호출 구조로 되어 있다. 따라서 레코드의 개수 n이 2의 거듭제곱이라고 가정하고 순환 호출의 깊이가 얼마나 되는지를 분석해 보자. n=2^3인 경우에는 부분배열의 크기가 2^3 -> 2^2 -> 2^1 -> 2^0  순으로 줄어들어 순환 호출의 길이가 3임을 알 수 있다. 따라서 n=2^k라고 하면 부분배열의 크기는 2^k -> 2^k-1 ...2^0이 되어 순환 호출의 깊이는 k 이때 k = log2n
