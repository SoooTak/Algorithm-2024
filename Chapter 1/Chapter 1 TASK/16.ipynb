#문제 16_맵

다음과 같은 맵이 있다. 물음에 답하라.

(1) 이 맵을 그래프 형태로 변환해보라. 하나의 영역은 정점이 되고 인접한 영역들 사이에는 간선이 있어야 한다.

(2) 이 맵에서 인접한 영역에 같은 색이 칠해지지 않도록 색을 칠하는 방법을 설명하라.

다른 정점과 가장 많이 인접한 B의 색을 먼저 정하고, 그 다음 많이 인접한 F, E, 그 다음 인접한 C, D, G, 마지막으로 A의 색을 정하면 다음과 같이 3가지 색으로 칠할 수 있다.

(3) 위 문제에서 가장 작은 수의 색을 사용해 색칠하는 방법을 구하라.

B와 맞닿아 있는 A, D, E에서 A와 E는 D를 사이에 두고 B와 이어져 있기에 2가지 색 이하로 칠할 수는 없다. 따라서 3가지 색으로 칠하는 것이 가장 적은 수의 색을 사용하는 방법이다..
